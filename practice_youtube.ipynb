```python
import openpyxl
import pandas
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
!ls "/content/drive/My Drive"

wb = openpyxl.load_workbook('/content/drive/MyDrive/us_youtube_trending.xlsx') #load_workbook - access an Excel file in python
sheet1 = wb['Sheet1']
data = sheet1.values
columns = next(data)[0:]
data_frame = pandas.DataFrame(data, columns = columns)
```

```python
#delete NaN records in video_id column
data_frame.dropna(axis = 0, subset = ['video_id'])
#axis = 0: delete the rows contain the null value
#axis = 1: delete the columns contain null value
```

```python
#show all the duplicate records in video_id column
print(data_frame[data_frame.duplicated(subset = ['video_id'], keep = False)])
#keep = False: all the duplicate records
#keep = First: all the duplicate records except the first record(unique)
#keep = Last: all the duplicate records except the last record(unique)
```

```python
#delete the duplicate records in video_id except the lastest trending date
data_frame_sorted = data_frame.sort_values('trending_date')
data_frame_sorted.drop_duplicates(subset = ['video_id'], keep = 'last')
```

```python
#check the data types of column in data_frame, change trending_date and publish_time into datetime type
data_frame['trending_date'] = pandas.to_datetime(data_frame['trending_date'])
data_frame['publish_time'] = pandas.to_datetime(data_frame['publish_time'])
print(data_frame.info())
```

```python
#descriptive statistic
print(data_frame.describe)
print(data_frame.corr())
```

```python
#sort data by channel_title column (A-Z) and views column (DESC)
data_frame.sort_values(by = ['channel_title', 'views'], ascending = [True, False])
```
